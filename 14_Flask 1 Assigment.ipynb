**Q3. What is App routing in Flask? Why do we use app routes?**

In Flask, app routing refers to the process of defining URL routes for your web application and associating them with specific Python functions. These routes determine how the application responds to different URLs requested by clients. <br>
Use of App Routes
1. URL Mapping: App routes define how URLs are mapped to specific views or functions in  Flask application. They provide a clean and organized way to handle different requests.
2. Code Organization: Routes help organize your code by associating related functionality with specific URLs making codebase more maintainable and easier to understand.
3. Dynamic Routing: Flask's support for dynamic routes allows you to create flexible routes that can handle variable data in the URL, such as user IDs, product names, or category names.
4. HTTP Methods: Routes can be restricted to specific HTTP methods (GET, POST, etc.), ensuring that the appropriate actions are taken for each type of request.

The url_for() function is used to build a URL to the specific function dynamically. The first argument is the name of the specified function, and then we can pass any number of keyword argument corresponding to the variable part of the URL. This function is useful in the sense that we can avoid hard-coding the URLs into the templates by dynamically building them using this function. 
This is particularly useful to avoid hardcoding URLs in your application, making it more maintainable and adaptable to changes.
