# Web Scraping

**Q1. What is Web Scraping? Why is it Used? Give three areas where Web Scraping is used to get data.**

Web Scraping is a process used to retrieve large amount of data from different websites using automated scripts. Web scraping uses crawler and scraper. Crawler follows the links in the web pages and information retrieval from those pages is done by scraper. The unstructured data is retrieved from the web and converted to structutred data for further analysis.<br>
Web Scraping is used in:
1. Sentiment Analysis: To understand the overall feedback of customers towards the product.
2. Price Monitoring: To place the products available at an optimum price range to maximize the profit.
3. Market Research: To Analyze consumer preference trends and behaviours in product selection.

**Q2. What are the different methods used for Web Scraping?**

1. HTML Parsing: HTML parsing involves the use of JavaScript to target a linear or nested HTML page. It is a powerful and fast method for extracting text and links.
2. DOM Parsing: The Document Object Model (DOM) defines the structure, style and content of an XML file. Scrapers typically use a DOM parser to view the structure of web pages in depth.
3. Vertical Aggregation:  These are data harvesting platforms that can be run on the cloud and are used to automatically generate and monitor bots for certain verticals with minimal human intervention
4. XPath: XPath is short for XML Path Language, which is a query language for XML documents. XML documents have tree-like structures, so scrapers can use XPath to navigate through them by selecting nodes according to various parameters.
5. Google Sheets:Scarpers can use the IMPORTXML function in Sheets to scrape from a website, which is useful if they want to extract a specific pattern or data from the website.
