Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Plots which can be plotted in Pyplot include line plot, scatter plot, bar plot, histogram, contour plot, violin plot, pie plot, box plot, surface plot etc.

**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (4,3))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.scatter(x, y, c='#1FD165')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

With subplot function in one figure, multiple plots can be drawn.

```python
plt.figure(figsize = (10,6))
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])
plt.subplot(2,2,1)
plt.title("First Plot")
plt.plot(x1, y1)

x2 = np.array([0,1,2,3,4,5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.title("Second Plot")
plt.plot(x2, y2)

x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.title("Third Plot")
plt.plot(x3, y3)

x4 = np.array([0,1,2,3,4,5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.title("Fourth Plot")
plt.plot(x4, y4)
plt.suptitle("Sub plots ")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize = (5,4))
plt.bar(company, profit)
plt.show()
```

```python
plt.figure(figsize = (5,4))
plt.barh(company, profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Boxplot is a standardized way of displaying the dataset based on the five number summary: minimum, maximum, sample median, first quartile and third quartile

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2])
plt.show()
```
