**Q1. Explain GET and POST methods.**

GET: This method is used to retrieve data from the server. <br>
POST: This method enables to users to send data across server

**Q2. Why is request used in Flask?**

The request, in Flask, is an object that contains all the data sent from client to server. This data can be recovered using POST/GET methods.

**Q3. Why is redirect() used in Flask?**

Redirect is used to send the user to particular URL with the status code.

**Q4. What are templates in Flask? Why is the render_template() function used?**

Templates are files that contains static data as well as placeholders for dynamic data. Template is rendered with specific data to produce a final document. render_template() is used to generate output from a template file based on Jinja2 engine that is found in the application's template folder.
